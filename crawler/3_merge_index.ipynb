```python
import pandas as pd

# Load the two CSV files into DataFrames
file1 = '/Users/alan/11711/nlp-from-scratch-assignment/data/crawled/sublink_file_name_url_mapping.csv'  # The larger CSV file
file2 = '/Users/alan/11711/nlp-from-scratch-assignment/data/crawled/sublink_file_name_url_mapping_cleaned_valid_url.csv'  # The smaller CSV file

df1 = pd.read_csv(file1)  # First CSV with more entries
df2 = pd.read_csv(file2)  # Second CSV with fewer entries

# Ensure the index of df1 is kept as a column (optional step)
df1 = df1.reset_index()  # This adds the original index as a column

# Perform an inner right join based on the 'value' column
# 'right' means keeping all rows from df2 (smaller CSV) that match in df1
result = pd.merge(df1, df2, on='Value', how='inner', suffixes=('_file1', '_file2'))

# Optionally, drop duplicate columns from df2 (if needed) after merging
# Example: if there are columns with the same name that you only want from file1
# result = result.loc[:, ~result.columns.duplicated()]

# Save the result to a new CSV file if needed
result.to_csv('/Users/alan/11711/nlp-from-scratch-assignment/data/crawled/sublink_file_name_url_mapping_filtered.csv', index=False)

print("Join operation complete. Result saved to output CSV.")
```

```python
import os
import pandas as pd

# Set the paths to the CSV file and the directory containing the text files
csv_file_path = '/Users/alan/11711/nlp-from-scratch-assignment/data/crawled/sublink_file_name_url_mapping_filtered.csv'  # Path to the CSV file
text_files_directory = '/Users/alan/11711/nlp-from-scratch-assignment/data/crawled/crawled_sublink_text_data/'  # Path to the directory containing {index}.txt files

# Read the CSV file and extract the column of indexes
df = pd.read_csv(csv_file_path)
index_column = 'Index_file1'

# Convert the column of indexes to a set for faster lookup
index_set = set(df[index_column].astype(str))  # Ensure the indexes are strings to match file names

# Iterate over all files in the directory
for file_name in os.listdir(text_files_directory):
    # Check if the file ends with .txt
    if file_name.endswith('.txt'):
        # Extract the index part of the file name (e.g., '123' from '123.txt')
        file_index = os.path.splitext(file_name)[0]
        
        # Check if this index is not in the CSV file's index set
        if file_index not in index_set:
            # Construct the full file path
            file_path = os.path.join(text_files_directory, file_name)
            
            # Delete the file
            os.remove(file_path)
            print(f"Deleted: {file_path}")

print("Cleanup complete.")
```

```python
import os
import shutil

def copy_files_with_prefix(src_dir, dst_dir, prefix):
    # Ensure the destination directory exists
    if not os.path.exists(dst_dir):
        os.makedirs(dst_dir)

    # Iterate over all files in the source directory
    for file_name in os.listdir(src_dir):
        # Construct full file path in the source directory
        full_file_path = os.path.join(src_dir, file_name)

        # Only copy files (ignore directories)
        if os.path.isfile(full_file_path):
            # Add the prefix to the file name
            new_file_name = prefix + file_name

            # Construct the full path for the new file in the destination directory
            dst_file_path = os.path.join(dst_dir, new_file_name)

            # Copy the file to the destination directory
            shutil.copy(full_file_path, dst_file_path)
            print(f"Copied: {full_file_path} to {dst_file_path}")

if __name__ == "__main__":
    # Set the source directory, destination directory, and prefix
    src_directory = "/Users/alan/11711/nlp-from-scratch-assignment/data/crawled/crawled_sublink_text_data/"  # Replace with the path to your source directory
    dst_directory = "/Users/alan/11711/nlp-from-scratch-assignment/data/crawled/crawled_all/"  # Replace with the path to your destination directory
    prefix = "sublink_"  # Replace with the desired prefix

    # Call the function to copy files with the prefix
    copy_files_with_prefix(src_directory, dst_directory, prefix)
```
